## Validación cruzada

En la [validación cruzada](https://es.wikipedia.org/wiki/Validaci%C3%B3n_cruzada) de K iteraciones o K-fold cross-validation los datos de muestra se dividen en K subconjuntos.

* Uno de los subconjuntos se utiliza como datos de prueba y el resto (K-1) como datos de entrenamiento.
* El proceso de validación cruzada es repetido durante k iteraciones, con cada uno de los posibles subconjuntos de datos de prueba.
* Finalmente se realiza la media aritmética de los resultados de cada iteración para obtener un único resultado.



```python
# carga el conjunto de datos
X, y = datasets.load_iris(return_X_y=True)
```

```python
# librerías
from sklearn.tree import DecisionTreeClassifier
```

```python
# modelo para clasificar: árbol de decisiones
modelo = DecisionTreeClassifier(random_state=7)
```

```python
# librerías
from sklearn.model_selection import KFold, cross_val_score
```

```python
# validación cruzada
k_folds = KFold(n_splits = 5)
```

```python
# metricas
scores = cross_val_score(modelo, X, y, scoring='accuracy', cv = k_folds)
```

```python
# muestra los resultados
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Std Dev CV Score: ", scores.std())
print("Number of CV Scores used in Average: ", len(scores))
```
