##Eliminar columnas

A menudo, descubrirás que no todas las categorías de datos de un conjunto de datos le resultan útiles.

##Ordenar campos

Hasta ahora, hemos eliminado columnas innecesarias y hemos cambiado el índice de nuestro ```DataFrame``` a algo más sensato.

* En esta sección, limpiaremos columnas específicas y las llevaremos a un formato uniforme para comprender mejor el conjunto de datos y reforzar la coherencia.
* En particular, limpiaremos la fecha de publicación y el lugar de publicación.

* Tras la inspección, todos los tipos de datos son actualmente el tipo de objeto, que es más o menos análogo a ```str``` en Python nativo.

* Encapsula cualquier campo que no se pueda ajustar claramente como datos numéricos o categóricos.
* Esto tiene sentido ya que estamos trabajando con datos que inicialmente son un montón de cadenas desordenadas.

````python
df.dtypes.value_counts()
````

Un campo donde tiene sentido aplicar un valor numérico es la fecha de publicación para que podamos hacer cálculos en el futuro.

````python
df.loc[1905:, 'Date of Publication'].head(10)
````

Un libro en particular sólo puede tener una fecha de publicación. Por lo tanto, debemos hacer lo siguiente:

* Elimine las fechas adicionales entre corchetes, siempre que estén presentes: 1879 $[1878]$.
* Convertir rangos de fechas a su "fecha de inicio", siempre que esté presente: 1860-63; 1839, 38-54.
* Elimine por completo las fechas de las que no estamos seguros y reemplácelas con NaN de NumPy: [1897?].
* Convierta la cadena nan al valor NaN de NumPy

* Sintetizando estos patrones, podemos aprovechar una única expresión regular para extraer el año de publicación.

Sintetizando estos patrones, podemos aprovechar una única expresión regular para extraer el año de publicación:

```regex = r'^(\d{4})'```

La expresión regular anterior está destinada a encontrar cuatro dígitos al comienzo de una cadena, lo cual es suficiente para nuestro caso.
* Lo anterior es una cadena sin formato.
* Lo que significa que una barra invertida ya no es un carácter de escape.
* Lo cual es una práctica estándar con expresiones regulares.

* ```\d``` representa cualquier dígito y ```{4}``` repite esta regla cuatro veces.
* El carácter ```^``` coincide con el inicio de una cadena.
* Los paréntesis indican un grupo de captura, que indica a Pandas que queremos extraer esa parte de la expresión regular.
* Queremos que ```^``` evite casos en los que ```[ ``` comience con la cadena ```)```.


Si no estás familiarizado con las expresiones regulares, puedes inspeccionar la expresión anterior en "[regex101.com](https://regex101.com/)" y aprender todo sobre las expresiones regulares.

````python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
````

Técnicamente, esta columna todavía tiene tipo de objeto, pero podemos obtener fácilmente su versión numérica con ```pd.to_numeric```.

````python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
````

Esto da como resultado que falte aproximadamente uno de cada diez valores, lo cual es un pequeño precio a pagar por poder hacer cálculos sobre los valores válidos restantes.


````python
df['Date of Publication'].isnull().sum() / len(df)
````

##Combinando métodos

Combinando métodos ```str``` con NumPy para limpiar columnas.

* Anteriomente utilizamos: ```df['Fecha de publicación'].str```.
* Este atributo es una forma de acceder a operaciones rápidas de cadenas en Pandas.
* Que imitan en gran medida operaciones en cadenas nativas de Python.
* O expresiones regulares compiladas, como ```.split()```, ```.replace()``` y ```.capitalize()```.

* Para limpiar el campo ```Lugar de publicación```.
* Podemos combinar los métodos pandas ```str``` con la función ```np.where``` de NumPy.
* Que es básicamente una forma vectorizada de la macro ```IF()``` de Excel.
* Tiene la forma de: ```np.where(condition, then, else)```.


````python
df['Place of Publication'].head(10)
````

Vemos que en algunas filas el lugar de publicación está rodeado de otra información innecesaria.
* Si miráramos más valores, veríamos que este es el caso solo para algunas filas cuyo lugar de publicación es "Londres" u "Oxford".

````python
df.loc[4157862]
````

````python
df.loc[4159587]
````

Estos dos libros se publicaron en el mismo lugar, pero uno tiene guiones en el nombre del lugar mientras que el otro no.

* Para limpiar esta columna de un solo barrido, podemos usar ```str.contains()``` para obtener una máscara booleana.

````python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
````

````python
oxford = pub.str.contains('Oxford')
````

Aquí, la función ```np.where``` se llama en una estructura anidada, siendo la condición una serie de booleanos obtenidos con ```str.contains()```.
* El método ```contains()``` funciona de manera similar a la palabra clave incorporada utilizada para encontrar la aparición de una entidad en un iterable (o subcadena en una cadena).

* El reemplazo a utilizar es una cadena que representa nuestro lugar de publicación deseado.
* También reemplazamos guiones con un espacio con ```str.replace()``` y los reasignamos a la columna en nuestro ```DataFrame```.

* Aunque hay más datos sucios en este conjunto de datos, por ahora solo discutiremos estas dos columnas.


````python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
````
