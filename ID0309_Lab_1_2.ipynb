##Paso 2: Combinar el conjunto de datos

##Paso 3: Reconstruir los datos faltantes

Para buscar y completar los datos que faltan en el conjunto de datos, usaremos otra función.

Hay varias formas de encontrar los valores nulos si están presentes en el conjunto de datos.

* Usando la función ```isnull()```.
* Usando la función ```isna()```.
* Usando la función ```isna().any()```.
* Usando la función ```isna(). sum()```.
* Usando la función ```isna().any().sum()```.

**Usando la función ```isnull()```**

* Esta función proporciona el valor Booleano para el conjunto de datos completo para saber si algún valor nulo está presente o no.

````python
data.isnull()
````

**Usando la función ```isna()```**

* Esta función Esto es lo mismo que la función ```isnull()```.
* Y proporciona el mismo resultado.

````python
data.isna()
````

**Usando la función ```isna().any()```**

* Esta función también proporciona un valor Booleano.
* Si hay algún valor nulo presente o no.
* Pero proporciona resultados en columnas.
* No en formato tabular.

````python
data.isna().any()
````

**Usando la función ```isna().sum()```**

* Esta función proporciona la suma de los valores nulos preestablecidos en el conjunto de datos en columnas.

````python
data.isna().sum()
````

**Usando la función ```isna().any().sum()```**

* Esta función proporciona resultados en un valor único si hay algún nulo presente o no.

````python
data.isna().any().sum()
````

No hay valores nulos presentes en nuestro conjunto de datos.


En caso de que hubiere valores nulos preestablecidos, podemos llenar esos lugares con cualquier otro valor usando la función ```fillna()``` de ```DataFrame```:


  ```DataFrame_name.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)```

##Paso 4: Estandarización y Normalización

La estandarización y normalización de datos es una práctica común en el aprendizaje automático.

* La estandarización es otra técnica de escala en la que los valores se centran alrededor de la media con una desviación estándar unitaria.
* Esto significa que la media del atributo se vuelve cero y la distribución resultante tiene una desviación estándar unitaria.
* La normalización es una técnica de escala en la que los valores se cambian y se reescalan para que terminen oscilando entre 0 y 1.
* También se conoce como escala [Min-Max](https://www.analyticsvidhya.com/blog/2020/04/feature-scaling-machine-learning-normalization-standardization/).

* Este paso no es necesario para el conjunto de datos que estamos utilizando.

##Paso 5: Eliminar datos duplicados

* No es necesario duplicar valores en el análisis de datos.
* Estos valores solo afectan la precisión y eficiencia del resultado del análisis.
* Para encontrar valores duplicados en el conjunto de datos, usaremos una función de marco de datos simple, es decir, ```duplicado()```.

````python
data.duplicated()
````

Esta función también proporciona valores booleanos para valores duplicados en el conjunto de datos.
* Como podemos ver, el conjunto de datos no contiene ningún valor duplicado.
* Si un conjunto de datos contiene valores duplicados, se puede eliminar usando la función ```drop_duplicates()```.
* La siguiente es la sintaxis de esta función:

  ```DataFrame_name.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)```
